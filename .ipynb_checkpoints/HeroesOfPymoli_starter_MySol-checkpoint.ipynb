```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# Raw data file
file_to_load = "Resources/purchase_data.csv"

# Read purchasing file and store into pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
# Find the total number of players by evaluating the number of values in a single column
player_count = purchase_data["Purchase ID"].count()
player_count
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Finds the number of unique items in the "Item ID" column
unique_items = len(purchase_data["Item ID"].unique())

# Finds the average price of values in the "Price" column
avg_price = purchase_data["Price"].mean()

# Finds the total revenue by summing up values in the "Price" column
total_revenue = purchase_data["Price"].sum()

# Create a DataFrame displaying above values
purchase_summary = pd.DataFrame({"Number of Unique Items": [unique_items],
                                "Average Price": avg_price,
                                "Number of Purchases": player_count,
                                "Total Revenue": total_revenue})

# Convert "Average Price" and "Total Revenue" columns to currency format
purchase_summary["Average Price"] = purchase_summary["Average Price"].map("${:,.2f}".format)
purchase_summary["Total Revenue"] = purchase_summary["Total Revenue"].map("${:,.2f}".format)

# Display DataFrame
purchase_summary
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Create a groupby object that groups by Gender
gender_group = purchase_data.groupby("Gender")

# Create a DataFrame based on the counts of the groupby object
gender_counts = gender_group.count()
gender_counts
```

```python
# Use loc method to find the total number of male players
male_count = gender_counts.loc["Male","Purchase ID"]

# Calculate the percent of total players who are male
percent_male = (male_count/player_count) * 100

# Use loc method to find the total number of female players
female_count = gender_counts.loc["Female","Purchase ID"]

# Calculate the percent of total players who are female
percent_female = (female_count/player_count) * 100

# Use loc method to find the total number of players who identify as "Other / Non-Disclosed"
other_count = gender_counts.loc["Other / Non-Disclosed","Purchase ID"]

# Calculate the percent of total players who are "Other / Non-Disclosed"
percent_other = (other_count/player_count) * 100

# Create a DataFrame displaying above values
gender_display = pd.DataFrame({"Gender": ["Male", "Female", "Other / Non-Disclosed"],
                              "Percentage of Players": [percent_male, percent_female, percent_other],
                              "Total Count": [male_count, female_count, other_count]})

# Convert "Percentage of Players" column to percentage format
gender_display["Percentage of Players"] = gender_display["Percentage of Players"].map("{:,.2f}%".format)

# Change the index to be the "Gender" column
gender_display.set_index("Gender", inplace=True)

# Display the DataFrame
gender_display
```

* Categorize the existing players using the age bins. Hint: use pd.cut()

* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
# Establish bins for ages
age_bins = [0, 9.90, 14.90, 19.90, 24.90, 29.90, 34.90, 39.90, 99999]
group_names = ["<10", "10-14", "15-19", "20-24", "25-29", "30-34", "35-39", "40+"]

# Categorize the values of the"Age" column into bins
cut_ages = pd.cut(purchase_data["Age"], age_bins, labels=group_names)

# Create a DataFrame based on value_counts of age categories
age_ranges = pd.DataFrame({"Total Count": cut_ages.value_counts()})

# Sort the DataFrame in ascending order by the index such that the lower age categories are first
age_ranges.sort_index(axis=0, inplace=True)

# Create a new column in the age_ranges DataFrame that calculates the fraction of total players in that age range
age_ranges["Percentage of Players"] = (age_ranges["Total Count"]/player_count) * 100

# Convert the new column to percent format
age_ranges["Percentage of Players"] = age_ranges["Percentage of Players"].map("{:,.2f}%".format)

# Display the DataFrame
age_ranges
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Group the original DataFrame by screenname
top_spenders = purchase_data.groupby("SN")

# Create a Series with "SN" as the index and purchase count as the value
spender_purchase_count = top_spenders["Price"].count()

# Find the total amount spent for each screenname
spender_total = top_spenders["Price"].sum()

# Calculate the average price spent for each screenname by dividing their total amount spent by the number of purchases
spender_avg_price = spender_total/spender_purchase_count

# Create a DataFrame with calculated values
spender_df = pd.DataFrame({"Purchase Count": spender_purchase_count,
                          "Average Purchase Price": spender_avg_price,
                          "Total Purchase Value": spender_total})

# Sort DataFrame by "Total Purchase Value" column in descending order
spender_df.sort_values(by=['Total Purchase Value'], ascending=False, inplace=True)

# Convert "Average Purchase Price" and "Total Purchase Value" columns to curry format
spender_df["Average Purchase Price"] = spender_df["Average Purchase Price"].map("${:,.2f}".format)
spender_df["Total Purchase Value"] = spender_df["Total Purchase Value"].map("${:,.2f}".format)

# Display first 5 rows of DataFrame
spender_df.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Create a DataFrame from a subsection of original DataFrame
item_df = purchase_data.loc[:, ['Item ID', 'Item Name', 'Price']]

# Create a groupby object by grouping the item DataFrame by "Item ID" and "Item Name" columns
item_group = item_df.groupby(["Item ID", "Item Name"])

# Find the number of times the item was purchased
item_count = item_group['Price'].count()

# Find the total amount spent on the item
item_total = item_group['Price'].sum()

# Find the price of the individual item by dividing the total amount spent by the purchase count
item_price = item_total/item_count



# Create a DataFrame displaying the above values
item_summary = pd.DataFrame({"Purchase Count": item_count,
                             "Item Price": item_price,
                             "Total Purchase Value": item_total})

# Sort the DataFrame by "Purchase Count" column in descending order 
item_sorted = item_summary.sort_values(by=['Purchase Count'], ascending=False)

# Convert the "Item Price" and "Total Purchase Value" columns to currency format
item_sorted["Item Price"] = item_sorted["Item Price"].map("${:,.2f}".format)
item_sorted["Total Purchase Value"] = item_sorted["Total Purchase Value"].map("${:,.2f}".format)

# Display the first 5 rows of the DataFrame
item_sorted.head()
```
